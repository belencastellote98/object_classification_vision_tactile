```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
objects = ["test_full_bottle", "test_empty_bottle", "test_full_can", "test_empty_can", "test_hard_ball", "test_tennis_ball"] 
objects = ["tomato_good", "tomato_bad"] 
```

```python
writer = pd.ExcelWriter('tomato_.xlsx', engine='xlsxwriter')
for name in objects:
    df = pd.read_csv(name+".csv")
    df_notime = df.drop(columns=["Time"])
    values=pd.DataFrame(columns = df_notime.columns)

    count=0
    dif_0 = 0
    # last_update = pd.DataFrame(columns=df.columns)
    max=0
    for i in range(len(df)):
        if len(values)>=100:
            break
        df_help = df_notime[df_notime.index==i]
        if df_help.mean(axis=1)[i]!=0:
            count=0
            dif_0 += 1
            possible = df_help
            if possible.sum(axis=1)[i]>max:
                max = possible.sum(axis=1)[i]
                last_update = df_help
        elif df_help.mean(axis=1)[i]==0 and max!=0:
            count+=1
            
            if count > 8:
                if dif_0>8:
                    values = pd.concat([values, last_update])
                    max=0
                    count = 0
                    dif_0=0
                else:
                    dif_0=0
            

        
    locals()["values_"+name]=values.copy()
            
    values.to_excel(writer, sheet_name=name)
writer.save()
```
